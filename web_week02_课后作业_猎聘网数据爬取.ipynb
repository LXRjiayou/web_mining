```python
import pandas as pd
from requests_html import HTMLSession

payload = {
    "key":"网页设计"
}
session = HTMLSession()
r = session.get("https://www.liepin.com/zhaopin/", params=payload)
print(r.status_code)
r.html.html  
```

```python
print(r.text)
```

```python
with open ("网页设计.html", encoding = "utf8", mode = "w") as fp:
    fp.write(r.html.html)
```

```python
with open ("网页设计.html", encoding = "utf8", mode = "r") as fp:
    html_load = fp.read()
```

```python
import requests_html
from requests_html import soup_parse
parsed = requests_html.soup_parse(html_load)
after = requests_html.soup_parse(html_load)
```

```python
jobs_name = r.html.xpath('//div[@class="job-info"]/h3/a')
for i in jobs_name:
    print(i.text)  # 获得工作标题
    print(i.absolute_links)  # 获得工作链接
```

```python
jobs_salary = r.html.xpath('//div[@class="job-info"]/p/@title')
print(jobs_salary)
```

```python
salary = r.html.xpath('//span[@class="text-warning"]/text()')
print(salary)
```

```python
edu = r.html.xpath('//span[@class="edu"]/text()')
print(edu)
```

```python
excises = r.html.xpath('//p[@class="condition clearfix"]/span[3]/text()')
print(excises)
```

```python
location = r.html.xpath('//p[@class="condition clearfix"]/a/text()')
print(location)
```

```python
location_url = r.html.xpath('//p[@class="condition clearfix"]/a/@href')
print(location_url)
```

```python
time = r.html.xpath('//p[@class="time-info clearfix"]/time/@title')
print(time)
```

```python
company_name = r.html.xpath('//p[@class="company-name"]/a/text()')
company_url = r.html.xpath('//p[@class="company-name"]/a/@href')
print(company_name)
print(company_url)
```

```python
df = pd.DataFrame({
    "职位名称":r.html.xpath('//div[@class="job-info"]/h3/@title'),
    "每月薪水":r.html.xpath('//span[@class="text-warning"]/text()'),
    "职位-url":r.html.xpath('//div[@class="job-info"]/h3/a/@href'),
    "公司名称":r.html.xpath('//p[@class="company-name"]/a/text()'),
    "公司-url":r.html.xpath('//p[@class="company-name"]/a/@href'),
    "最低学历":r.html.xpath('//span[@class="edu"]/text()'),
    "发布日期":r.html.xpath('//p[@class="time-info clearfix"]/time/@title'),
    "所需经验":r.html.xpath('//p[@class="condition clearfix"]/span[3]/text()'),
})
df
```
